## Load in Data

```python
df = pd.read_csv(r"C:\Users\om22285\Coding Dojo\Week 18\Coding Dojo\Applying Advanced Transformations (Core)\Applying-Advanced-Transformations--Core-\Data\superhero_info - superhero_info.csv")

# Create a copy to work with
df_1 = df.copy()

# display the head
df_1.head()
```

```python
df_1.info()
```

```python
# Split the Hero|Publisher column into two
## adding expand=True
df_1['Hero|Publisher'].str.split('|',expand=True)
```

```python
## save the 2 new columns into the dataframe
df_1[['Hero','Publisher']] = df_1['Hero|Publisher'].str.split('|',expand=True)
df_1.head(2)
```

```python
# Drop the original column 
df_1 = df_1.drop(columns= ['Hero|Publisher'])
df_1.head(2)
```

## Split the measurements column

```python
# Remove brackets
df_1['Measurements'] = df_1['Measurements'].str.strip('{}')

# perform loop to replace characters
# Create list of characters to be replaced
char_replace = ["'", ":","cm", "kg","Height", "Weight", " "]
for char in char_replace:
    df_1['Measurements'] = df_1['Measurements'].str.replace(char,'',regex=False)
    
## save the 2 new columns into the dataframe
df_1[['Height','Weight']] = df_1['Measurements'].str.split(',',expand=True)
df_1.head(2)
```

```python
# Drop the original column
df_1 = df_1.drop(columns='Measurements')

# convert columns Height & Weight to numeric
df_1['Weight'] = df_1['Weight'].astype('float')
df_1['Height'] = df_1['Height'].astype('float')
df_1.head()
```

## Load in the Second Set of data

```python
df2 = pd.read_csv(r"C:\Users\om22285\Coding Dojo\Week 18\Coding Dojo\Applying Advanced Transformations (Core)\Applying-Advanced-Transformations--Core-\Data\superhero_powers - superhero_powers.csv")

# Create a copy to work with
df_2 = df2.copy()

# display the head
df_2.head()
```

```python
# Create new column and convert original column to list format
df_2['Powers2'] = df_2['Powers'].to_list()

# Have to split the string firt but don't expand (create individual columns yet)
## exploding the column of lists
df_2['Powers'] = df_2['Powers'].str.split(',', expand=False)

exploded = df_2.explode('Powers')
#exploded[['name','transactions','transactions_split']].head(5)
```

```python
# view the new colulmn
df_2['Powers2']
```

```python
## saving the unique values from the exploded column
cols_to_make = exploded['Powers'].dropna().unique()

# col_list
col_list = cols_to_make.tolist()
cols_to_make
```

```python
# convert column back to string
df_2['Powers2'] = df_2['Powers2'].astype('string')
```

```python
# Perform For Loop to get 
for col in cols_to_make:
    df_2[col] = df_2['Powers2'].str.contains(col) #.isin(col_list) 
df_2.head()
```

```python
# Drop the columns
df_2 = df_2.drop(columns=['Powers', 'Powers2'])
```

## Merge the two tables

```python
## example merge with our 2 dataframes
df_final = pd.merge(df_1, df_2, left_on='Hero', right_on='hero_names', how='inner')
df_final.head(3)
```

```python
# Compare average weight of heros with super speed vs those without
df_final.groupby('Super Speed').Weight.mean().round(2)
```

```python
# Average height of heros from each publisher
df_final.groupby('Publisher').Height.mean().round(2)
```
